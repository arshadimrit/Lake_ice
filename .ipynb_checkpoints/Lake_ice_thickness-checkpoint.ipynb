# Add CRU data to lake files

```python
%%time
cru_df = pd.read_csv('/mnt/c/Users/imrit/Downloads/CRU_files/Processed/cru_mean_temp_cld_prec_1901_2020_formatted.csv')
cru_df = cru_df[['long', 'lat', 'year', 
                 'mean_cld_dec', 'mean_cld_jan', 'mean_cld_feb', 'mean_cld_mar', 
                 'mean_temp_dec', 'mean_temp_jan', 'mean_temp_feb', 'mean_temp_mar', 
                 'prec_dec', 'prec_jan', 'prec_feb', 'prec_mar']]
cru_df['mean_temp_djfm'] = cru_df[['mean_temp_dec', 'mean_temp_jan', 'mean_temp_feb', 'mean_temp_mar']].mean(axis=1)
cru_df['cld_djfm'] = cru_df[['mean_cld_dec', 'mean_cld_jan', 'mean_cld_feb', 'mean_cld_mar']].mean(axis=1)
cru_df['prec_djfm'] = cru_df[['prec_dec', 'prec_jan', 'prec_feb', 'prec_mar']].mean(axis=1)

display(cru_df.head())
```

```python
oscillations_df = pd.read_csv('/mnt/c/Users/imrit/Downloads/Sapna_work/Ice_thickness/Data/climate_oscillations_1837_2018.csv')
oscillations_df = oscillations_df[['PDO mean A-DEC' ,'QBO mean A-DEC', 'SOI mean A-DEC', 'NINO3 mean A-DEC', 'AO mean A-DEC', 
                                   'AMO mean A-DEC', 'NAO DJFM', 'global_merged_co2_annual (ppm)', 'ENSO paleo A-DEC', 
                                   'Daily Total Sunspot Number mean A-DEC', 'NH', 'SH', 'year']]
oscillations_df.rename(columns={'PDO mean A-DEC': 'PDO',
                                'QBO mean A-DEC': 'QBO', 
                                'SOI mean A-DEC': 'SOI', 
                                'NINO3 mean A-DEC': 'NINO3', 
                                'AO mean A-DEC': 'AO', 
                                'AMO mean A-DEC': 'AMO', 
                                'NAO DJFM': 'NAO', 
                                'global_merged_co2_annual (ppm)': 'CO2', 
                                'ENSO paleo A-DEC': 'ENSO', 
                                'Daily Total Sunspot Number mean A-DEC': 'Sunspot', 
                                'NH': 'NH', 
                                'SH': 'SH'}, inplace=True)
display(oscillations_df.head())
```

```python
path = '/mnt/c/Users/imrit/Downloads/Sapna_work/Ice_thickness/Data/Lakes_lat_long/'

lat_longs = []

files = [file for file in os.listdir(path)]

progress_bar = tqdm(files)

for file in progress_bar:
    progress_bar.set_description("Processing %s" % file)
    df = pd.read_csv(f'{path}{file}')
    
    lake_name = file.split('.csv')[0]
    latitude = df['Latitude'].mean()
    longitude = df['Longitude'].mean()
    
    df['waterbody_name'] = file.split('.csv')[0]
    df['Latitude'].fillna(df['Latitude'].mean(), inplace=True)
    df['Longitude'].fillna(df['Longitude'].mean(), inplace=True)
    try:
        df['Year'] = df['Year'].str.split('-').str.get(0).astype(int)
    except:
        df['Year'] = df['Year']
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    
    df['latgrid'] = np.round((df['latitude'] + 0.25) * 2) / 2.-0.25
    df['longrid'] = np.round((df['longitude'] + 0.25) * 2) / 2.-0.25
    
#     # Add Cru
    df_cru = df.merge(cru_df, left_on=['latgrid','longrid', 'year'], 
                       right_on=['lat','long', 'year'], how='left').drop(['lat', 'long', 'latgrid', 'longrid'], axis=1)
    
    # Add oscillations
    
    df_cru_oscillations = df_cru.merge(oscillations_df, on='year', how='left')
    
    df_cru_oscillations.to_csv(f'/mnt/c/Users/imrit/Downloads/Sapna_work/Ice_thickness/Data/Lakes_climate/{file}', 
                               index=False)
```

```python
path = '/mnt/c/Users/imrit/Downloads/Sapna_work/Ice_thickness/Data/Lakes_climate/'

y_columns =['maximum_ice_thickness_(cm)']
x_columns = ['mean_temp_djfm', 'cld_djfm', 'prec_djfm',
             'PDO', 'QBO', 'SOI', 'NINO3', 'AO', 'AMO', 'NAO', 'Sunspot', 'CO2']
results_columns = ['lake', 'numb_features', 'x_data', 'y_data', 'adj_r2_start_year', 
                   'r2', 'r2_adj', 'pvalues', 'coefs', 'aic', 'bic']

for file in os.listdir(path):
    lake_name = file.split('.')[0]
    print(lake_name)
    df = pd.read_csv(f'{path}{file}')
    
    best_subset_linear_regress = [] 
    
    Y = df['maximum_ice_thickness_(cm)']
    X = df[x_columns]
    start_year_data = df['year']
    
    if Y.isna().any():
        continue
        
    else:
        for k in trange(1, len(X.columns) + 1, desc = 'Loop...'):
            for combo in itertools.combinations(X.columns, k):
                X_2 = X[list(combo)]
                X_2 = X_2.dropna()
                X_2 = sm.add_constant(X_2)
                Y_2 = Y[-len(X_2):]
                Y_2 = list(Y_2)
                
                x_start_year = start_year_data[-len(X_2):]
                x_start_year = list(x_start_year)                

                results = sm.OLS(Y_2, X_2).fit()
                
                df = pd.DataFrame(columns=results_columns)
                df = df.append({'lake': lake_name, 
                                'numb_features': len(combo), 
                                'x_data': combo, 
                                'y_data': 'ice_thickness',
                                'r2': results.rsquared, 
                                'r2_adj': results.rsquared_adj, 
                                'pvalues':  results.pvalues.values, 
                                'coefs': results.params.values, 
                                'aic': results.aic, 
                                'bic': results.bic,
                                'log_likelihood': results.llf,
                                'f_pvalue': results.f_pvalue, 
                                'fvalue': results.fvalue, 
                                'model': results.model
                           }, ignore_index=True)
                best_subset_linear_regress.append(df)
        lake_linear_regress_subsets = pd.concat(best_subset_linear_regress).sort_values(by=['r2_adj'], 
                                                                                        ascending=True)
        lake_linear_regress_subsets.to_csv(f'/mnt/c/Users/imrit/Downloads/Sapna_work/Ice_thickness/Analysis/Multi_lin_reg/{lake_name}_linear_regress_subsets.csv', 
                                           index=False)
```
