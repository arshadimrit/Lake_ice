# Data filtering - November 11 2019

```python
full_lake = pd.read_csv('/mnt/c/Users/mimri/Downloads/All_lakes_ts_v3.0_sapna.csv')
full_lake.replace('-999', np.nan, inplace=True)
display(full_lake.head())
print(len(full_lake))
```

```python
unique_lakes_first_year = (full_lake.dropna(subset=['start_year', 'froze', 'ice_off_1']).
                          drop_duplicates(['lakecode'], keep=('first')))
unique_lakes_last_year = (full_lake.dropna(subset=['start_year', 'froze', 'ice_off_1']).
                          drop_duplicates(['lakecode'], keep=('last')))

first_last_lakes = pd.concat([unique_lakes_first_year, unique_lakes_last_year])
```

```python
less_1900_unique_lakes_first_year = unique_lakes_first_year[unique_lakes_first_year['start_year'] < 1900]
display(less_1900_unique_lakes_first_year.head())
print(len(less_1900_unique_lakes_first_year))

lakecodes_before_1900 = list(less_1900_unique_lakes_first_year['lakecode'])
print(len(lakecodes_before_1900))
```

```python
lakes_before_1900 = full_lake[full_lake['lakecode'].isin(lakecodes_before_1900)]
lakes_before_1900 = lakes_before_1900[lakes_before_1900['lakecode'] != 'ARAI1']
display(lakes_before_1900.head())
```

```python
#lakes_before_1900.to_csv('/mnt/c/Users/mimri/Downloads/lakes_before_1900.csv')
```

```python
first_last_lakes_before_1900 = first_last_lakes[first_last_lakes['lakecode'].isin(lakecodes_before_1900)]
first_last_lakes_before_1900 = first_last_lakes_before_1900[first_last_lakes_before_1900['lakecode'] != 'ARAI1']
display(first_last_lakes_before_1900.head())
print(len(first_last_lakes_before_1900))
```

```python
years_difference = []

for lakecode in lakecodes_before_1900[1:]:
    df = first_last_lakes_before_1900[first_last_lakes_before_1900['lakecode'] == lakecode]
    first_year = np.array(df.iloc[0]['start_year'])
    last_year = np.array(df.iloc[-1]['start_year'])
    lake_name = np.array(df.iloc[0]['lake'])
    contributor = np.array(df.iloc[0]['Contributor'])
    new_df = pd.DataFrame(np.array([[lakecode,lake_name, contributor, first_year, last_year]]), 
                          columns=['lakecode', 'lake', 'contributor','first_year', 'last_year'])
    new_df['first_year'] = new_df['first_year'].astype('int', inplace=True)
    new_df['last_year'] = new_df['last_year'].astype('int', inplace=True)
    years_difference.append(new_df)

first_last_year_df = pd.concat(years_difference)
first_last_year_df['year_difference_counts'] = first_last_year_df['last_year'] - first_last_year_df['first_year']
display(first_last_year_df.head())
```

```python
lake_before_1900_row_counts = lakes_before_1900.dropna(subset=['start_year', 'froze', 'ice_off_1'])['lakecode'].value_counts().rename_axis('lakecode').reset_index(name='counts_no_na')
display(lake_before_1900_row_counts.head())
```

```python
lake_before_1900_all_counts = pd.merge(first_last_year_df, lake_before_1900_row_counts, on='lakecode')
lake_before_1900_all_counts['difference_rows_counts']= lake_before_1900_all_counts['year_difference_counts'] - lake_before_1900_all_counts['counts_no_na']
display(lake_before_1900_all_counts.head(10))
print(len(lake_before_1900_all_counts))
```

```python
lakes_before_1900_less_15_missing = lake_before_1900_all_counts[lake_before_1900_all_counts['difference_rows_counts'] < 15]
# lakes_before_1900_less_15_missing.sort('first_year')
display(lakes_before_1900_less_15_missing)
print(len(lakes_before_1900_less_15_missing))
```
