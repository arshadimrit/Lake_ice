```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as graph
import seaborn as sns
import os
```

```python
df = pd.read_csv('/mnt/c/Users/imrit/Downloads/IceFreeLakes_v1.3.1.csv')
display(df.head())
print(df.columns)
```

```python
proportion_data = []
for lake_name in df['lake'].unique():
    df_lake = df[df['lake'] == lake_name]
    start_year, end_year, len_data = df_lake['start_year'].iloc[0], df_lake['start_year'].iloc[-1], len(df_lake)
    no_freeze = len(df_lake[df_lake['FrozeYN'] == 'N'])
    prop_no_freeze = no_freeze/len_data *100
    proportion_data.append([lake_name, start_year, end_year, len_data, no_freeze, prop_no_freeze])
```

```python
proportion_df = pd.DataFrame(proportion_data)
proportion_df.rename(columns= {0: 'lake', 1: 'start_year_date', 2: 'end_year_date', 3: 'length_data', 
                               4: 'number_no_freeze', 5: 'proportion_no_freeze'}, inplace=True)
proportion_df = proportion_df.round(2)
proportion_df.head()
```

```python
df_grouped = pd.DataFrame(df.groupby('lake').mean())
df_grouped = df_grouped.reset_index().round(2)
display(df_grouped)
```

```python
merged_columns = ['lake', 'Latitude', 'Longitude', 'Hylak_id', 'Lake_type',
       'Grand_id', 'Lake_area', 'Shore_len', 'Shore_dev', 'Vol_total',
       'Vol_res', 'Vol_src', 'Depth_avg', 'Dis_avg', 'Res_time', 'Elevation',
       'Slope_100', 'Wshd_area', 'LIAG_mean_depth',
       'LIAG_surface_area', 'LIAG_shoreline', 'start_year_date', 'end_year_date', 'length_data', 'number_no_freeze',
       'proportion_no_freeze']
```

```python
merged_df = pd.merge(df_grouped, proportion_df)
merged_df = merged_df[merged_columns]
display(merged_df.head())
merged_df.to_csv('/mnt/c/Users/imrit/Downloads/Sapna_sharma_lab/Permanent_ice_loss/Lake_characteristics.csv', index=False)
```
