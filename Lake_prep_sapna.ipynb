```python
# df = pd.read_csv('/mnt/c/Users/imrit/Downloads/Sapna_work/Prep/Switzerland/Lakes/pffaf_testing.csv')
path = '/mnt/c/Users/imrit/Downloads/Sapna_work/Prep/Switzerland/Lakes_v2_may_25/'

for file in os.listdir(path):
    lake_name = file.split('_')[0]
    print(f"Processing Lake {lake_name}")
    df = pd.read_csv(f"{path}{file}")
    percentage_missing = 1

    # lake_name = file.split('_')[0]
    df.rename(columns={'Frozen': 'froze', 'ice_on': 'ice_on_pre', 'ice_off': 'ice_off_pre', 'year': 'start_year'}, inplace=True)
    df['Country'] = 'Switzerland'
    df['contributor'] = 'Harrie-Jan Hendricks Franssen'
    df['lakecode'] = ""
    df['source'] = ""
    df['lake'] = lake_name
#     df['lake'] = ""
#     df['Longitude'] = ""
#     df["Latitude"] = ""
    df['Updated Year'] = ""
    
#     df['froze'] = np.where(df['start_year'] == 1948, 'N', df['froze'])

    missing_ice_on_year_conditions = (df['ice_on_pre'].isnull()) & (df['froze'].isin(['Y', '-999']))
    missing_ice_off_year_conditions = (df['ice_off_pre'].isnull()) & (df['froze'].isin(['Y', '-999']))

    df['ice_off_pre'] = df['ice_off_pre'].str.replace("Mrt", "Mar")
    df['ice_on_pre'] = df['ice_on_pre'].str.replace("Mrt", "Mar")

    df['ice_on_date'] = df['ice_on_pre'].str.extract('([0-9]\w{0,} ...)', expand=True)
    df['ice_off_date'] = df['ice_off_pre'].str.extract('([0-9]\w{0,} ...)', expand=True)

    df['ice_on_month'] = df['ice_on_date'].str.extract('([A-Z]\w{0,})', expand=True)
    df['ice_off_month'] = df['ice_off_date'].str.extract('([A-Z]\w{0,})', expand=True)
        
    df['ice_on'] = np.where(df['ice_on_month'].isin(['Nov', 'Dec']), 
                                 pd.to_datetime((df['start_year']).map(str) + '-' + df['ice_on_date']), 
                                 pd.to_datetime((df['start_year']+1).map(str) + '-' + df['ice_on_date']))
    df['ice_off'] = np.where(df['ice_off_month'].isin(['Nov', 'Dec']), 
                                 pd.to_datetime((df['start_year']).map(str) + '-' + df['ice_off_date']), 
                                 pd.to_datetime((df['start_year']+1).map(str) + '-' + df['ice_off_date']))

    df['ice_on_doy'] = (pd.to_datetime(df['ice_on']) - pd.to_datetime((df['start_year']).map(str) + '-' + 'Dec 31')).dt.days
    df['ice_off_doy'] = (pd.to_datetime(df['ice_off']) - pd.to_datetime((df['start_year']).map(str) + '-' + 'Dec 31')).dt.days
    df['duration'] = np.where(df['froze'] == 'N', 0, df['ice_off_doy'] - df['ice_on_doy'])


    df['Missing Year_IceON'] = np.where(missing_ice_on_year_conditions, 'Yes', 'No')
    df['Missing Year_IceOFF'] = np.where(missing_ice_off_year_conditions, 'Yes', 'No')

#     for col in ['ice_on_doy', 'ice_off_doy']:
    for col in ['ice_on_doy']:

        frac_missing= df[col].isnull().sum() / len(df)
        if frac_missing > percentage_missing:
            continue
        else:
            df[col] = np.where(((df[col].isnull()) & (df['froze'] == 'Y')), 
                               df[col].mean(), df[col])
            df[col] = np.where(df['froze'] == 'N', np.round(df[col].max()) if col == 'ice_on_doy' else df[col].min(), df[col])

    df = df[column_names]

#     display(df.head(), df.tail())
#     df.to_csv(f'/mnt/c/Users/imrit/Downloads/Sapna_work/Prep/Switzerland/Final_lakes_v2_may_25/{lake_name}.csv', index=False)
#     df.to_csv('/mnt/c/Users/imrit/Desktop/test.csv')
```

## Get time series length and number of missing years

```python
file_path = '/mnt/c/Users/imrit/Downloads/Sapna_work/Data_google_drive/April_22/'
lakes_data_length = []

for file in os.listdir(file_path):
    df = pd.read_csv(f'{file_path}{file}')
    df['missing_year'] = np.where((df['Missing Year_IceON'] == 'Yes') & (df['Missing Year_IceOFF'] == 'Yes') & (df['froze'] == 'Y'), 1, 0)
    length_data = len(df)
    number_missing_years = df['missing_year'].sum()
    lakes_data_length.append([file.split('.')[0], df['start_year'].iloc[0], df['start_year'].iloc[-1], length_data, 
                              number_missing_years])

length_years_df = pd.DataFrame(lakes_data_length)
length_years_df.rename(columns={0: 'lake', 1: 'start_year', 2: 'last_start_year', 
                                3: 'number_years', 4: 'number_missing_years'}, inplace=True)
# length_years_df.to_csv('/mnt/c/Users/imrit/Downloads/Sapna_work/Data_google_drive/Length_data_lakes_may_5_2020.csv', index=False)
display(length_years_df.head())
```
